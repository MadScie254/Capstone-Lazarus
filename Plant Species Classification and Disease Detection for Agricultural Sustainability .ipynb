## Step 2: Data Preprocessing

In this step here are the key tasks:

1. **Data Splitting**: Split the dataset into training, validation, and test sets. The typical split ratio is around 70-80% for training, 10-15% for validation, and 10-15% for testing.

```python
def get_dataset_partitions_tf(ds, train_split=0.8, test_split=0.1, val_split=0.1, shuffle=True, shuffle_size=1000):
    ds_size = len(ds)
    if shuffle:
        ds = ds.shuffle(shuffle_size, seed=12)
    train_size = int(train_split* ds_size)
    val_size = int(val_split* ds_size)
    
    train_ds =  ds.take(train_size)
    
    val_ds =  ds.skip(train_size).take(val_size)
    test_ds =  ds.skip(train_size).skip(val_size)
    
    return train_ds, test_ds, val_ds
```

## Oversampling and Undersampling

**Oversampling** - Randomly replicate samples from "minority classes"

```python
import os
import random
import shutil

# Define path
dataset_path = r'data'

# Target path
target_class = "Corn_(maize)___Northern_Leaf_Blight"

# desired no of samples
desired_num_samples = 2000 # Adjust when needed

# create new folder to store the oversampled data
oversampled_path = os.path.join(dataset_path, f"{target_class}_oversampled")
os.makedirs(oversampled_path, exist_ok=True)

# loop through elements
target_class_path = os.path.join(dataset_path, target_class)
target_class_images = os.listdir(target_class_path)

while len(target_class_images) < desired_num_samples:
    img_file = random.choice(target_class_images)
    img_path =os.path.join(target_class_path, img_file)
    new_img_path = os.path.join(oversampled_path, img_file)
    shutil.copyfile(img_path, new_img_path)
    target_class_images.append(img_file)
    
print(f"Oversampling completed for {target_class} class")
```

Oversampling is used in the context of handling class imbalances in a dataset, where one or more classes have significantly fewer samples compared to other classes. Justifications and explanations for using oversampling in your project:

1. **Improve Model Performance**: Class imbalance can lead to biased model training, where the model may perform well on the majority class but poorly on the minority class. By oversampling the minority class, you provide the model with more examples to learn from, improving its ability to recognize and correctly classify the minority class.

2. **Prevent Biased Predictions**: Without oversampling, the model may be biased toward the majority class because it's seen more frequently during training. This can result in high accuracy but low recall for the minority class, meaning that the model may fail to correctly identify instances of the minority class.

3. **Capture Minority Class Patterns**: Oversampling ensures that the model encounters a broader range of examples from the minority class. This helps the model learn the subtle patterns, features, and variations within the minority class, making it more capable of distinguishing between different instances of that class.

4. **Avoid Data Loss**: In cases of severe class imbalance, undersampling the majority class to match the minority class may result in a significant loss of data and information. Oversampling allows you to balance the dataset without discarding valuable data.

5. **Generalization**: Oversampling can help improve the generalization ability of the model. By presenting the model with a more balanced dataset, you reduce the risk of overfitting to the majority class, which can lead to better model performance on unseen data.

## Undersampling

```python
import os
import random
import shutil

# Define the path to the dataset folder
dataset_path = r'data'

# Define the class (subfolder) for which you want to apply undersampling (e.g., "maize")
target_class = "Corn_(maize)___Northern_Leaf_Blight"

# Define the desired number of samples (matching the minority class)
desired_num_samples = 500  # Adjust as needed

# Create a new folder to store undersampled images
undersampled_path = os.path.join(dataset_path, f"{target_class}_undersampled")
os.makedirs(undersampled_path, exist_ok=True)

# Loop through the images in the target class and randomly select a subset
target_class_path = os.path.join(dataset_path, target_class)
target_class_images = os.listdir(target_class_path)

while len(target_class_images) > desired_num_samples:
    img_file = random.choice(target_class_images)
    img_path = os.path.join(target_class_path, img_file)
    new_img_path = os.path.join(undersampled_path, img_file)
    shutil.copyfile(img_path, new_img_path)
    target_class_images.remove(img_file)

print(f"Undersampling completed for {target_class} class.")
```

Undersampling is a technique used in addressing class imbalances, where you reduce the number of samples in the majority class to match the number of samples in the minority class. Justification and deployment:

1. **Imbalanced Data**: When you have a highly imbalanced dataset, where one class significantly outweighs the others in terms of the number of samples. In such cases, machine learning models may have a bias toward the majority class, leading to poor classification performance for minority classes.

2. **Resource Constraints**: Undersampling can be an appropriate choice when you have limitations on computational resources, time, or budget. Collecting additional data for the minority class may not be feasible due to these constraints.

3. **Data Quality**: If the minority class data is of high quality and represents real-world scenarios accurately, undersampling can help preserve this valuable data. Oversampling methods, on the other hand, generate synthetic data that may not fully capture the nuances of the original class.

4. **Risk of Overfitting**: Oversampling can lead to overfitting, where the model becomes too specialized in predicting the minority class. Undersampling reduces the risk of overfitting by focusing on a smaller subset of the majority class, making it harder for the model to memorize the data.

5. **Interpretability**: In some cases, it might be more interpretable and understandable to work with a balanced dataset, as it simplifies the training process and model evaluation. Undersampling helps achieve this balance.

However, it's essential to be cautious when using undersampling, as it does come with potential drawbacks:

- **Loss of Information**: Undersampling reduces the available data for training, which may lead to a loss of valuable information present in the majority class.

- **Reduced Variability**: With fewer samples in the majority class, the model may not capture the full variability of that class.

- **Risk of Bias**: Undersampling may not always be suitable for every problem. It assumes that the majority class instances are redundant or similar, which may not be the case in all scenarios.

- **Impact on Model Complexity**: Depending on the chosen undersampling strategy, it may affect the overall complexity of the model and its ability to generalize.
